Before we take our next step there's some other things we need to do first. First is looping over the rows and taking the average values for all the different age groups then making this a new column. This is just to see if there are any patterns we might miss not looking at the whole picture.

```python
age_groups = ['18-24', '25-49', '50-64', '65+']
reso['All Ages'] = reso.apply(lambda row: row[age_groups].mean(), axis=1)
young = reso.sort_values(by='18-24',ascending=True)
middle = reso.sort_values(by='25-49',ascending=True)
old = reso.sort_values(by='50-64',ascending=True)
senior = reso.sort_values(by='65+',ascending=True)
average = reso.sort_values(by='All Ages',ascending=True)
```

```python
plt.barh(average['Resolution'], average['All Ages'])
plt.xlabel('Percentage')
plt.ylabel('Resolution')
plt.title('New Years Resolution for All Ages')
plt.show()
```

```python
plt.barh(young['Resolution'], young['18-24'])
plt.xlabel('Percentage')
plt.ylabel('Resolution')
plt.title('New Years Resolution for 18-24 Age Group')
plt.show()
```

```python
plt.barh(middle['Resolution'],middle['25-49'])
plt.xlabel('Percentage')
plt.ylabel('Resolution')
plt.title('New Years Resolution for 25-49 Age Group')
plt.show()
```

```python
plt.barh(old['Resolution'],old['50-64'])
plt.xlabel('Percentage')
plt.ylabel('Resolution')
plt.title('New Years Resolution for 50-64 Age Group')
plt.show()
```

```python
plt.barh(senior['Resolution'],senior['65+'])
plt.xlabel('Percentage')
plt.ylabel('Resolution')
plt.title('New Years Resolution for 65+ Age Group')
plt.show()
```

Here we see one constant about all of the age groups. Everyone wants to focus on being healthier whether that's ffrom eating better or it's from being more active and getting more excercise. the 18-24 age group and the 50-64 age groups seemed to be the one's that focused on these goals the most at over 60% of people said this is what their resolution was. The least common goal was to quit smoking as in every age group this was right around the bottom and never coming above 10%. Raising money for charity also follows the same guidelines.
